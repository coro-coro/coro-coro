```python
import pandas as pd
import numpy as np
import pyecharts 
```

```python
print("pyecharts version : ", pyecharts.__version__)
```

```python
data = pd.read_csv('patient_data.csv', engine='python') #옵션 써야 한글 읽힘
```

```python
data.head()
```

```python
data = data.fillna("")
```

```python
data2 = data.groupby(['date', 'degree']).size()
print(data2)
```

```python
data2 = data2.reset_index()
data2.head()
```

```python
data2.columns = ['date', 'degree', 'count']
data2.head()
```

```python
data2['cum_count'] = list(data2['count'].cumsum())
data2.head()
```

```python
count1 = []
for x in range(data2.shape[0]):
    if data2.loc[x, 'degree'] == "1차":
        count1.append(data2.loc[x,'count'] * 1)
    else: count1.append(data2.loc[x,'count'] * 0)
data2['cum_count1'] = list(np.cumsum(count1))
```

```python
count2 = []
for x in range(data2.shape[0]):
    if data2.loc[x, 'degree'] == "2차":
        count2.append(data2.loc[x,'count'] * 1)
    else: count2.append(data2.loc[x,'count'] * 0)
data2['cum_count2'] = list(np.cumsum(count2))
```

```python
count3 = []
for x in range(data2.shape[0]):
    if data2.loc[x, 'degree'] == "3차":
        count3.append(data2.loc[x,'count'] * 1)
    else: count3.append(data2.loc[x,'count'] * 0)
data2['cum_count3'] = list(np.cumsum(count3))
```

```python
count_unknown = []
for x in range(data2.shape[0]):
    if data2.loc[x, 'degree'] == "확인 중":
        count_unknown.append(data2.loc[x,'count'] * 1)
    else: count_unknown.append(data2.loc[x,'count'] * 0)
data2['cum_unknown'] = list(np.cumsum(count_unknown))
```

```python
print(data2)
```

```python
data_fin = data2.groupby('date').apply(max, axis=0)
data_fin = data_fin.drop(['date', 'degree', 'count'], axis=1)
print(data_fin)
```

```python
# 누적확진자
import pyecharts.options as opts
from pyecharts.charts import Line, Page
from pyecharts.commons.utils import JsCode
from pyecharts.faker import Collector, Faker
```

```python
#ver_grid
cum_patient = (
    Line()
    .add_xaxis(list(data_fin.index))
    .add_yaxis(
        "", list(data_fin['cum_count']),label_opts=opts.LabelOpts(is_show=True, color=['black'], font_size=11),
        is_selected=False)
    .add_yaxis(
        "확인 중", list(data_fin['cum_unknown']), areastyle_opts=opts.AreaStyleOpts(opacity=1), stack="stack1",
        label_opts=opts.LabelOpts(is_show=False)
    )
    .add_yaxis(
        "3차", list(data_fin['cum_count3']), areastyle_opts=opts.AreaStyleOpts(opacity=1), stack="stack1",
        label_opts=opts.LabelOpts(is_show=False)
    )
    .add_yaxis(
        "2차", list(data_fin['cum_count2']), areastyle_opts=opts.AreaStyleOpts(opacity=1), stack="stack1",
        label_opts=opts.LabelOpts(is_show=False)
    )
    .add_yaxis(
        "1차", list(data_fin['cum_count1']), areastyle_opts=opts.AreaStyleOpts(opacity=1), stack="stack1",
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_colors(["maroon", "gray", "orangered",  "firebrick", "maroon"])
    .set_global_opts(title_opts=opts.TitleOpts(title=""),
                    xaxis_opts = opts.AxisOpts(type_ = "time", split_number=15),
                    yaxis_opts = opts.AxisOpts(interval = 5))
)
cum_patient.render_notebook()
```
