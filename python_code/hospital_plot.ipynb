```python
import pandas as pd
import numpy as np
import pyecharts
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
data = pd.read_csv('patient_data.csv', engine='python')
```

```python
data.head()
```

```python
hospital = list(np.sort(data.hospital.unique()))
hospital_value = list(data.hospital.value_counts().sort_index())
```

```python
hospital
```

```python
hospital_value
```

```python
hospital_bar = (
        Bar()
        .add_xaxis(hospital)
        .add_yaxis("확진자 수(명)", hospital_value, color="#BDBDBD")
        .set_global_opts(title_opts=opts.TitleOpts(title="격리병원"))
        .set_series_opts(label_opts=opts.LabelOpts(position="right", color=["black"]))
        .reversal_axis()
)
```

```python
hospital_bar.render()
```

```python
hospital_bar.render_notebook()
```
