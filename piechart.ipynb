```python
import pandas as pd
import numpy as np
import pyecharts 
```

```python
print("pyecharts version : ", pyecharts.__version__)
```

```python
data = pd.read_csv('patient_data.csv', engine='python') #옵션 써야 한글 읽힘
```

```python
data.head()
```

```python
data = data.fillna("")
data.head()
```

```python
data.shape
```

```python
from pyecharts.charts import Pie
from pyecharts import options as opts
```

```python
# 감염차수 pie plot
degree = list(np.sort(data.degree.unique()))
degree_value = list(data.degree.value_counts().sort_index())

degree_plot = (
    Pie()
    .add("", [list(z) for z in zip(degree, degree_value)])
    .set_colors(["maroon", "firebrick", "orangered", "gray"])
    .set_global_opts(title_opts=opts.TitleOpts(title=""),
                    legend_opts=opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts=opts.LabelOpts(color=["black"], formatter="{b}:\n{d}%", font_size = 11)) #d:%, c:명수
)
degree_plot.render('chasu.html')
```

```python
# 상태별 pie chart
condition = list(np.sort(data.condition.unique()))
condition_value = list(data.condition.value_counts().sort_index())

condition_plot=(
    Pie()
    .add("", [list(z) for z in zip(condition, condition_value)])
    .set_colors(["cornflowerblue", "slategray", "#BDBDBD"])
    .set_global_opts(title_opts=opts.TitleOpts(title="상태(명)"),
                    legend_opts=opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts=opts.LabelOpts(color=["black"], formatter="{b}: {d}%", font_size = 11)) #d:%, c:명수
)
condition_plot.render('status.html')
```

```python
degree_plot.render_notebook()
```

```python
condition_plot.render_notebook()
```

```python
# 성별 pie chart

sex = list(np.sort(data.sex.unique()))
sex_value = list(data.sex.value_counts().sort_index())

sex_plot = (
    Pie()
    .add("", [list(z) for z in zip(sex, sex_value)])
    .set_colors(["cornflowerblue", "salmon","gray"])
    .set_global_opts(title_opts=opts.TitleOpts(title="성별(명)"),
                    legend_opts=opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts=opts.LabelOpts(color=["black"], formatter="{b}: {d}%", font_size = 11)) #d:%, c:명수
)
sex_plot.render('gend.html')
```

```python
sex_plot.render_notebook()
```

```python
# 감염유형별 pie chart
infection = list(np.sort(data.type.unique()))
infection_value = list(data.type.value_counts().sort_index())

infection_plot = (
    Pie()
    .add("", [list(z) for z in zip(infection, infection_value)])
    .set_colors(['#fc6472', '#f4b2a6', '#eccdb3', '#bcefd0', '#a1e8e4', 'gray', '#23c8b2', '#7f5a7c'])
    .set_global_opts(title_opts=opts.TitleOpts(title=""),
                    legend_opts=opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts=opts.LabelOpts(color=["black"], formatter="{b}:\n{d}%", font_size = 11)) #d:%, c:명수
)
infection_plot.render('how.html')
```

```python
infection_plot.render_notebook()
```

```python
# 국적별 pie chart
country = list(np.sort(data.country.unique()))
country_value = list(data.country.value_counts().sort_index())

country_plot = (
    Pie()
    .add("", [list(z) for z in zip(country, country_value)])
    .set_colors(['#b09880', '#f0d0c8', 'gray']) 
    .set_global_opts(title_opts=opts.TitleOpts(title="국적(명)"),
                    legend_opts=opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts=opts.LabelOpts(color=["black"], formatter="{b}: {d}%", font_size = 11)) #d:%, c:명수
)
country_plot.render('nat.html')
```

```python
country_plot.render_notebook()
```
