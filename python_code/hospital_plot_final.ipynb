```python
import pandas as pd
import numpy as np
import pyecharts
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
data = pd.read_csv('patient_data.csv', engine='python')
```

```python
hospital = list(np.sort(data.hospital.unique()))
hospital_value = list(data.hospital.value_counts().sort_index())
```

```python
hospital_bar = (
        Bar()
        .add_xaxis(hospital)
        .add_yaxis("확진자 수(명)", hospital_value, color="#BDBDBD")
        .set_global_opts(title_opts=opts.TitleOpts(title="격리병원"),
                         xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-13)),
                         datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")])
        .set_series_opts(label_opts=opts.LabelOpts(position="top", color=["black"]))
)
```
